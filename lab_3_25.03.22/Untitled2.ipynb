```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
disease=['appendicitis','Breast_Cancer_Wisconsin','hepatitis','indian_liver_patient']
alg=['alg1','alg2','alg3','alg4','alg5']
col=['Accuracy','Error','Sensitivity','Specificity']
algo=[0,0,0,0,0]
```

```python
class diseases:
    def __init__(self,disease_name):
        self.disease_name=disease_name
    def read(self):
        for i in range(0,5):
            algo[i]=pd.read_csv("C:\\Users\\Ujjwal\\Desktop\\nila\\Input\\%s\\%s\\AverageIter1000.csv"%(self.disease_name,alg[i]),nrows=100)
            
    def manuplate(self):
        for i in range(0,5):
            algo[i]=algo[i][algo[i].columns[0:4]]
            algo[i].columns=col
    def graph(self):
        #fig , ax=plt.subplots(nrows=2,ncols=2)
        for i in col:
            plt.figure()
            plt.title('%s_%s'%(self.disease_name,i))
            for j in range(0,5):
                sns.distplot(algo[j][i],kde=True,hist=False,label=alg[j]).set_ylabel("Density")
            plt.savefig('%s_%s.png'%(self.disease_name,i))
```

```python
a=[0,0,0,0]
for i in range(0,4):
    a[i]=diseases(disease[i])
    a[i].read()
    a[i].manuplate()
    a[i].graph()
```
